```python
now = datetime.now()

t1 = now.strftime("%H : %M : %S")
print('Time: ', t1)

t2 = now.strftime("%d : %m : %y, %H : %M : %S")
print('Date and time: ', t2)

t2 = now.strftime("%m : %d : %y, %H : %M : %S")
print('Date and time: ', t2)

print('Year: ', now.strftime('%y'))
print('Month: ', now.strftime('%m'))
print('Day: ', now.strftime('%d'))
```
